```python
from openai import OpenAI

[![Open In Colab](https://colab.research.google.com/assets/colab-badge.svg)](https://colab.research.google.com/github/nursnaaz/zero-to-genai-engineer/blob/main/05_Local_LLMs_and_API_Providers/notebooks/NB6_Databricks_Endpoint.ipynb)

import os

# To get a DATABRICKS_TOKEN, click the "Generate Access Token" button or follow https://docs.databricks.com/en/dev-tools/auth/pat.html
DATABRICKS_TOKEN = os.environ.get("DATABRICKS_TOKEN", "your-token-here")  # export DATABRICKS_TOKEN=... in terminal

client = OpenAI(
  api_key=DATABRICKS_TOKEN,
  base_url="https://dbc-9bf4f388-2fab.cloud.databricks.com/ai-gateway/mlflow/v1"
)

chat_completion = client.chat.completions.create(
  messages=[
    {"role": "user", "content": "Hello!"},
    {"role": "assistant", "content": "Hello! How can I assist you today?"},
    {"role": "user", "content": "What is Databricks?"},
  ],
  model="databricks-qwen35-122b-a10b",
  max_tokens=1024
)

print(chat_completion.choices[0].message.content)
```
