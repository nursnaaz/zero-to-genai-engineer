## 🎯 One-Line Summary
> At 175 billion parameters, GPT-3 performs competitive NLP tasks with just a **few examples in the prompt** — no gradient updates, no fine-tuning, no task-specific data.

---

## ❓ Problem It Solves
GPT-2 showed zero-shot learning is possible, but performance was still below fine-tuned models.  
Every real-world application still needed its own fine-tuned model.

**The question GPT-3 asked:**  
*If you give the model a few examples of a task inside the prompt (but still update no weights), does it perform as well as a fine-tuned model?*

---

## 🔧 How It Works

### In-Context Learning (ICL)
Three settings, all with **zero weight updates:**

```python
# Zero-shot: just describe the task
"Translate English to French: cat => "

# One-shot: one example
"Translate English to French:
dog => chien
cat => "

# Few-shot: a few examples
"Translate English to French:
dog => chien
cat => chat
house => "
```

The model reads the examples, infers the pattern, and continues.

### Scale
| | GPT-2 | GPT-3 |
|---|---|---|
| Parameters | 1.5B | **175B** |
| Training tokens | 40GB | **300B tokens** |
| Context window | 1024 | **2048 tokens** |
| Training cost | ~$50K | **~$4.6M** |

---

## 📊 Key Results
- Few-shot GPT-3 **matched fine-tuned BERT** on many benchmarks
- Generated news articles that humans couldn't distinguish from real ones 52% of the time
- Performed arithmetic, coding, translation — all from examples in the prompt
- Spawned the entire "prompt engineering" field

---

## 💡 Key Insights
1. **In-context learning is not learning.** No weights change. The model uses attention to "condition" on examples in its context window.
2. **Scale creates emergent capabilities.** Tasks the model couldn't do at 1B params suddenly appear at 175B — not gradual improvement, sudden jumps.
3. **The API model is born.** GPT-3 was never open-sourced — you access it via API. This is the foundation of OpenAI's business model.

---

## ⚠️ Limitations (the paper was honest about these)
- Still makes factual errors
- Context window limits few-shot examples to ~2000 tokens
- No memory between conversations
- Expensive to run (~$0.02 per 1K tokens initially)
- "Aligned to predict text, not to be helpful" → led to InstructGPT

---

## 📍 Where It Fits

```
GPT-2: zero-shot, scale to 1.5B
GPT-3: few-shot, scale to 175B  ← HERE
InstructGPT: same base, aligned with RLHF → ChatGPT
GPT-4: multimodal + further scale
```
