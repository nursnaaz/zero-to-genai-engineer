## 🎯 One-Line Summary
> BERT reads text **left AND right simultaneously** using masked language modelling, producing richer representations than GPT's left-to-right approach — but it cannot generate text.

---

## ❓ Problem It Solves
GPT reads text left-to-right only (causal). When predicting a word, it can only see what came before.

For **understanding** tasks (classification, Q&A, NER), you need to understand context from **both directions**.

Example: *"The bank can guarantee deposits will eventually cover future tuition costs."*  
Does "bank" mean a financial institution or a river bank?  
You need the right-side context ("deposits", "tuition") to know — GPT can't see that.

---

## 🔧 How It Works

### Architecture: Transformer Encoder (NOT Decoder)
- GPT uses the **decoder** (causal mask: only sees left context)
- BERT uses the **encoder** (no causal mask: sees all context in both directions)

### Two Pre-Training Objectives

**1. Masked Language Modelling (MLM)**
```
Input:  "The [MASK] sat on the mat."
Target: predict "cat"
```
15% of tokens are masked. BERT must predict them using context from both sides.

**2. Next Sentence Prediction (NSP)**
```
Sentence A: "The man went to the store."
Sentence B: "He bought a gallon of milk."  → IsNextSentence
Sentence B: "Penguins live in Antarctica." → NotNextSentence
```
Teaches the model to understand relationships between sentences.

---

## 📊 Key Results
- Set new state-of-the-art on **11 NLP benchmarks** at time of release
- GLUE score: 80.4% (previous best: 72.8%)
- SQuAD v1.1 F1: 93.2% (human performance: 91.2%)

---

## 💡 BERT vs GPT — The Key Difference

| | BERT | GPT |
|---|---|---|
| **Architecture** | Encoder | Decoder |
| **Direction** | Bidirectional | Left-to-right only |
| **Pre-training** | Mask + predict | Next token prediction |
| **Good at** | Understanding (classification, QA) | **Generation (writing text)** |
| **Can generate?** | ❌ No | ✅ Yes |
| **Use today** | Embeddings, search, classification | ChatGPT, Gemini, Claude |

---

## 💡 Key Insights
1. **Bidirectional context = better representations.** Seeing both sides of a word gives richer embeddings.
2. **Encoder ≠ Decoder.** This distinction determines whether a model can generate or only understand.
3. **BERT won NLP in 2018–2020.** Almost every NLP leaderboard was topped by BERT variants (RoBERTa, ALBERT, DistilBERT).
4. **GPT won the future.** Despite BERT's dominance, generative models (GPT family) ultimately became the paradigm for LLMs.

---

## 📍 Where It Fits

```
Transformer (2017) → BERT uses the encoder half
                   → GPT uses the decoder half

BERT (2018) → best for UNDERSTANDING tasks
GPT  (2018) → best for GENERATION tasks
Today       → decoder-only models dominate (GPT, Claude, Gemini, LLaMA)
```
