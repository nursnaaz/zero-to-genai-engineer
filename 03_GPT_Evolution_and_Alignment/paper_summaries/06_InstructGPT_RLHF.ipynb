## 🎯 One-Line Summary
> InstructGPT used Reinforcement Learning from Human Feedback (RLHF) to make GPT-3 follow instructions helpfully — this is the direct predecessor of ChatGPT.

---

## ❓ Problem It Solves
Raw GPT-3 is trained to predict the next token. This makes it very good at completing text but **not at following instructions**:

```
User: "Explain the moon landing to a 5-year-old."
GPT-3: "Explain the moon landing to a 5-year-old.
        Explain the moon landing to a 10-year-old.
        Explain the moon landing to a scientist."  # ← continues the pattern, doesn't answer
```

GPT-3 is "aligned with the training data" but not "aligned with what users actually want."

---

## 🔧 The RLHF Pipeline — 3 Steps

### Step 1: Supervised Fine-Tuning (SFT)
- Human labellers write ideal responses to prompts
- Fine-tune GPT-3 on these demonstrations
- Result: GPT-3-SFT — a model that tries to be helpful

### Step 2: Reward Model (RM)
- Show labellers pairs of GPT outputs for the same prompt
- Labellers rank which is better
- Train a separate model to predict the ranking
- Result: a "reward model" that scores any response for quality

### Step 3: Reinforcement Learning with PPO
- Use the reward model as a reward signal
- Fine-tune GPT-3-SFT with RL (PPO algorithm)
- The model learns to generate responses that score high on the reward model
- Result: InstructGPT

```
Human prompts
     ↓ Step 1
SFT model (demonstrations)
     ↓ Step 2
Reward Model (trained on human rankings)
     ↓ Step 3
PPO fine-tuning (maximise reward)
     ↓
InstructGPT ← this is what became ChatGPT
```

---

## 📊 Key Results
- **InstructGPT (1.3B params) preferred over GPT-3 (175B params)** by human labellers 85% of the time
- Smaller, aligned model > bigger, unaligned model
- Significantly fewer "hallucinations" and harmful outputs
- Better at following instructions, being concise, being honest about uncertainty

---

## 💡 Key Insights
1. **Alignment ≠ capability.** A smaller aligned model beats a bigger unaligned one on what users actually want.
2. **Human feedback is the missing ingredient.** Scale alone doesn't make a model helpful — you need human preference signal.
3. **RLHF is the recipe for ChatGPT.** This paper describes exactly how ChatGPT was built. GPT-4 uses the same approach with more scale and data.
4. **The "helpful, harmless, honest" target.** The labeller guidelines defined what "good" means — and that definition shapes what the model becomes.

---

## ⚠️ Limitations
- Labeller bias — different people rank differently
- "Alignment tax" — RLHF slightly hurts performance on standard benchmarks
- Reward hacking — model can learn to game the reward model
- PPO is complex and unstable → motivated DPO (see Paper 10)
