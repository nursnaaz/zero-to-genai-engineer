## 🎯 One-Line Summary
> Google's systematic study showing that using an AI to provide preference feedback (RLAIF) achieves **comparable results to human feedback (RLHF)** — validating AI-labelling as a scalable alternative.

---

## ❓ Problem It Solves
RLHF requires humans to compare model outputs and pick the better one — for every training example.  
This is:
- Slow (humans can only review so many pairs per day)
- Expensive (especially for expert feedback)
- Hard to scale as models get better and subtler differences are harder to judge

**The question RLAIF asked:**  
*Can we replace human raters entirely with an off-the-shelf LLM, and get the same quality alignment?*

---

## 🔧 How It Works

### Drop-in Replacement for Human Raters
RLHF pipeline:
```
Human looks at (Response A, Response B) → picks better one → trains reward model
```

RLAIF pipeline:
```
LLM  looks at (Response A, Response B) → picks better one → trains reward model
```

Everything else (reward model training, PPO fine-tuning) stays the same.

### The Annotator LLM
- Uses a prompted LLM (e.g. PaLM 2) as the rater
- Prompt includes a rubric for quality
- LLM outputs a preference score or ranking

---

## 📊 Key Results
- **RLAIF ≈ RLHF** on summarisation tasks (win rates vs SFT baseline approximately equal, ~71–73%; RLAIF slightly higher in reported results)
- Human raters preferred RLAIF-trained model outputs at similar rates to RLHF-trained
- RLAIF scales better — you can generate millions of AI preference labels cheaply
- Works best when the annotator LLM is significantly stronger than the model being trained

---

## 💡 Key Insights
1. **AI feedback is good enough.** For many tasks, an LLM rater produces alignment quality comparable to human raters.
2. **Scales with model capability.** As LLMs improve, RLAIF quality improves automatically.
3. **Cost reduction is massive.** Human annotations at scale cost millions; LLM annotations cost thousands.
4. **Independent validation of Constitutional AI.** Google confirmed what Anthropic had shown — AI feedback works.

---

## ⚠️ Limitations
- Annotator LLM can encode its own biases
- Not studied for safety/harmlessness (only helpfulness)
- Requires a capable enough annotator LLM — weak LLMs produce weak signal
- Circular risk: if the annotator has the same flaws as the model, you reinforce those flaws
