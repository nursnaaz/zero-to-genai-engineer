## 🎯 One-Line Summary
> A language model trained at sufficient scale can perform many tasks **without any fine-tuning** — just by reading the task description in the prompt.

---

## ❓ Problem It Solves
GPT-1 still needed fine-tuning on each task. That means:
- Labelled data per task
- A separate training run per task
- A separate model checkpoint per task

**The question GPT-2 asked:**  
*If the model is large enough and trained on diverse enough data, does it already "know" how to do tasks without being explicitly trained on them?*

---

## 🔧 How It Works

### Scale Everything
| | GPT-1 | GPT-2 |
|---|---|---|
| Parameters | 117M | **1.5B** |
| Layers | 12 | 48 |
| Context window | 512 tokens | **1024 tokens** |
| Training data | BooksCorpus (~800M tokens) | **WebText (40GB, ~8B tokens)** |

### Zero-Shot Learning
No fine-tuning. The task is described in the prompt itself:

```
Translate English to French:
cheese =>
```

The model has seen so much text that it learned translation as a side effect of learning to predict the next word.

---

## 📊 Key Results
- Achieved **state-of-the-art zero-shot** on 7 out of 8 language modelling benchmarks
- Generated coherent multi-paragraph text that was mistaken for human writing
- OpenAI initially **withheld the full model** due to misuse concerns — the first AI safety controversy

---

## 💡 Key Insights
1. **"Tasks can be described in natural language."** You don't need task-specific training data — just describe the task.
2. **Scale is a superpower.** More data + more parameters → emergent capabilities.
3. **The internet is a multitask dataset.** Web text contains instructions, Q&A, translation, code — a model that predicts all of it learns all of them.

---

## ⚠️ The Famous Controversy
OpenAI said the model was "too dangerous to release" and released it in stages over 9 months.  
Many in the research community criticised this as overhyped.  
This debate about responsible AI disclosure continues today.
