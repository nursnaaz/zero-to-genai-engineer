## 🎯 One-Line Summary
> Anthropic's foundational RLHF paper — the research behind Claude — showing how to train assistants that are simultaneously helpful, harmless, and honest (HHH), and revealing the fundamental tension between helpfulness and harmlessness.

---

## ❓ Problem It Solves
OpenAI's InstructGPT focused on "following instructions well."  
Anthropic asked a harder question: **How do you train a model that is helpful AND safe at the same time?**

The tension: Making a model more helpful often makes it more willing to assist with harmful requests.  
Making it more harmless often makes it overly cautious and useless.

---

## 🔧 How It Works

### The HHH Framework
Three properties the model should have simultaneously:

| Property | Meaning |
|---|---|
| **Helpful** | Gives useful, accurate, actionable responses |
| **Harmless** | Refuses or redirects genuinely harmful requests |
| **Honest** | Doesn't hallucinate; acknowledges uncertainty |

### RLHF — Same Pipeline as InstructGPT
1. Collect human preference data on (helpful, harmless) dimensions separately
2. Train separate reward models for helpfulness and harmlessness
3. Fine-tune with RL using a combined reward

### Key Addition: Preference Model (PM) Scores
Anthropic built **two separate reward models** — one for helpfulness, one for harmlessness — and studied the tension between them.

---

## 📊 Key Results
- Confirmed that RLHF produces significantly more helpful and harmless models
- **Discovered the helpfulness-harmlessness trade-off** — improving one hurts the other
- Showed that "red-teaming" (adversarial prompting) is essential for evaluation
- Published the HH dataset (~170K human preference pairs) — still widely used

---

## 💡 Key Insights
1. **Helpfulness and harmlessness conflict.** You can't just optimise both independently — you need to manage the trade-off explicitly.
2. **Red-teaming is essential.** You must actively try to break your model to know what it will do in production.
3. **Human labellers encode values.** Whoever writes the guidelines decides what the model believes is good behaviour.
4. **This is the foundation of Claude.** Anthropic's subsequent work (Constitutional AI, Claude 1/2/3) all builds on this paper's framework.
