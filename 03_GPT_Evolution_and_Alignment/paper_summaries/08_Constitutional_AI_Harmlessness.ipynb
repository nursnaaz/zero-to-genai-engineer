---

## 🎯 One-Line Summary
> Instead of paying humans to label harmful responses, Constitutional AI uses an AI to critique and revise its own outputs based on a written "constitution" — making alignment scalable without unlimited human labelling.

---

## ❓ Problem It Solves
Human feedback for harmlessness is:
- **Expensive** — trained labellers cost money per example
- **Slow** — doesn't scale with model capability
- **Inconsistent** — different labellers have different standards
- **Traumatising** — labellers see harmful content repeatedly

**The question Constitutional AI asked:**  
*Can an AI model critique and improve its own responses according to a set of written principles — without needing human feedback on every harmful example?*

---

## 🔧 How It Works — Two Stages

### Stage 1: Supervised Learning from AI Feedback (SL-CAI)
1. **Red-team the model** — prompt it to produce harmful responses
2. **Critique** — ask the same model: *"Is this response harmful? Why?"*
3. **Revise** — ask it to rewrite the response to be harmless
4. Repeat critique → revise loop multiple times
5. Fine-tune on the final revised responses

### Stage 2: Reinforcement Learning from AI Feedback (RLAIF)
1. Generate pairs of responses
2. Ask the AI to evaluate which is more harmless according to the constitution
3. Use these AI preferences to train a **preference model**
4. Fine-tune with RL using this AI-generated preference signal

### The Constitution
A written list of principles, for example:
> *"Choose the response that is least likely to contain harmful or unethical content."*
> *"Choose the response that is most supportive of human autonomy and individual freedom."*
> *"Choose the response that a thoughtful, senior Anthropic employee would be proud of."*

---

## 📊 Key Results
- Models trained with CAI were **as harmless as RLHF models** (trained with human labels)
- **Significantly more helpful** than RLHF-harmless models (avoided over-refusal)
- Much less human labelling required for the harmlessness component
- CAI models explained their refusals better than RLHF models

---

## 💡 Key Insights
1. **AI can evaluate AI.** A capable language model can judge whether a response is harmful — this is the RLAIF insight.
2. **Principles beat labels.** A short written constitution can replace thousands of human preference examples.
3. **Transparency.** The constitution is published — users can read what values the model is trained on.
4. **This is Claude.** Anthropic used CAI to build Claude 1, Claude 2, and Claude 3.
