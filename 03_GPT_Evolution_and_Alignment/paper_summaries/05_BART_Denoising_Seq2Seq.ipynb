## 🎯 One-Line Summary
> BART combines a **BERT-style bidirectional encoder** with a **GPT-style autoregressive decoder**, pre-trained by corrupting text and learning to reconstruct it — making it the best model of its era for summarisation.

---

## ❓ Problem It Solves
BERT is great at understanding but **can't generate**.  
GPT is great at generating but only sees left context.

Summarisation requires both:
- **Understanding** the input (full document — needs bidirectional context)
- **Generating** the output (fluent summary — needs autoregressive decoding)

Neither BERT nor GPT alone is optimal for this.

---

## 🔧 How It Works

### Architecture: Full Encoder-Decoder Transformer
```
[Corrupted Input]
       ↓
Bidirectional Encoder (BERT-style)
       ↓  cross-attention
Autoregressive Decoder (GPT-style)
       ↓
[Reconstructed Original Text]
```

### Pre-Training: Denoising
BART is trained to **reconstruct the original text** after it has been corrupted.  
Five corruption strategies tested:

| Corruption | Example |
|---|---|
| Token Masking | "The [MASK] sat on [MASK] mat" |
| Token Deletion | "The sat on mat" |
| Text Infilling | "The [MASK] mat" (multiple tokens → one mask) |
| Sentence Permutation | Shuffle sentence order |
| Document Rotation | Start from a random token |

**Best: Text Infilling + Sentence Permutation** (used in final model)

---

## 📊 Key Results
- **CNN/DailyMail summarisation**: ROUGE-1 of 44.16 — new state-of-the-art
- **ELI5 abstractive QA**: outperformed all previous models
- **Translation**: competitive with supervised baselines using only monolingual data
- SQuAD and MNLI also competitive with RoBERTa

---

## 💡 Key Insights
1. **The best of both worlds.** Encoder sees full document context; decoder generates fluently.
2. **Corruption strategy matters.** Text infilling (masking spans) is better than single-token masking.
3. **Pre-training task should match downstream task.** BART's reconstruction objective aligns well with summarisation and translation.

---

## 🗺️ The Encoder-Decoder Family

| Model | Architecture | Best for |
|---|---|---|
| BERT | Encoder only | Classification, Q&A |
| GPT | Decoder only | Open-ended generation |
| **BART** | Encoder + Decoder | **Summarisation, Translation** |
| T5 | Encoder + Decoder | "Text-to-Text" everything |
