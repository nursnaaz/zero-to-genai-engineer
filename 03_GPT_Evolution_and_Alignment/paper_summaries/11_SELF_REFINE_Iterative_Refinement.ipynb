## 🎯 One-Line Summary
> A single LLM can iteratively improve its own outputs by generating feedback on its own response and then revising — no training, no additional models, no human labels needed.

---

## ❓ Problem It Solves
LLM outputs are often good but not great on the first attempt.  
Traditional approaches to improvement require:
- Training a new model (expensive)
- Human feedback (slow)
- A separate critic/judge model (complex pipeline)

**The SELF-REFINE insight:**  
*The same model that generates the answer also knows enough to critique it and improve it.*  
Just like how a good writer reviews and revises their own drafts.

---

## 🔧 How It Works — 3-Step Loop

```
┌─────────────────────────────────────────────────────┐
│                                                     │
│  1. GENERATE   → Initial output (draft answer)      │
│       ↓                                             │
│  2. FEEDBACK   → Self-critique ("what's wrong?")    │
│       ↓                                             │
│  3. REFINE     → Improved output based on feedback  │
│       ↓                                             │
│  Repeat until stopping condition met                │
│                                                     │
└─────────────────────────────────────────────────────┘
```

### Example — Code Generation
```
Step 1 Generate:
  "def sort_list(lst): return sorted(lst)"
  
Step 2 Feedback (self-critique):
  "This doesn't handle None inputs. It also doesn't document 
   the function. The variable name is not descriptive."
   
Step 3 Refine:
  "def sort_list_ascending(lst: list) -> list:
       '''Sort a list in ascending order. Returns empty list for None.'''
       if lst is None: return []
       return sorted(lst)"
       
Step 4 Feedback: "Good improvement. Could add type checking..."
Step 5 Refine: ...
```

### Stopping Conditions
- Fixed number of iterations (e.g. 3)
- Feedback says "no changes needed"
- Quality score above threshold

---

## 📊 Key Results
Tested on 7 tasks: dialogue, code, math reasoning, acronym generation, story generation, sentiment reversal, constrained generation

- Improved over initial output on **most evaluated tasks** (math reasoning showed smallest gains)
- Average relative improvement of ~20% over initial generation (using GPT-3.5, as reported in paper)
- GPT-4 with SELF-REFINE outperformed GPT-4 baseline on most tasks
- No additional training data or human labels required

---

## 💡 Key Insights
1. **LLMs are better critics than generators.** It's easier for a model to spot problems than to generate perfectly on the first try.
2. **Iteration is free.** You don't need to train anything — just keep prompting the same model.
3. **This is how humans work.** First draft → review → revise → repeat. SELF-REFINE formalises this as a prompting strategy.
4. **Foundation for agentic AI.** Most modern AI agents (LangGraph, CrewAI) incorporate self-critique loops inspired by this work.

---

## ⚠️ Limitations
- Costs more tokens (multiple forward passes per answer)
- Can over-refine — the model sometimes makes good responses worse
- Quality depends on the model's ability to self-critique (weak models can't self-critique well)
- No guarantee of convergence
